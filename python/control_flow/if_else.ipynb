## 🔍 Concept Explanation
Python uses `if`, `elif`, and `else` gates to control program branching based on boolean conditions.

### 🌟 Why This Matters in Data Science:
Branching forms the basis of business logic, decision trees, and validation scripts.

### 🧠 Core Concepts & Mental Models:
- **Truthy vs Falsy:** Non-empty objects evaluate to True; empty structures and 0 evaluate to False.
- **Ternary Operator:** One-line conditional assignments: `value = A if condition else B`.

---

## 📝 Detailed Explanations
Branching in Python evaluates conditions sequentially. Under logical short-circuiting, operations stop as soon as the final truth value is guaranteed.



### 💻 Example 1: Standard if-elif-else branching


```python
score = 85
if score >= 90:
    print('Grade A')
elif score >= 80:
    print('Grade B')
else:
    print('Grade C')
```

### 💻 Example 2: Ternary conditional expression


```python
age = 20
status = 'Adult' if age >= 18 else 'Minor'
print('Status:', status)
```

### 💻 Example 3: Nested conditional branching


```python
num = 15
if num % 2 != 0:
    print('Odd')
    if num % 5 == 0:
        print('Divisible by 5')
```

### 💻 Example 4: Logical short-circuit evaluation in if gates


```python
def check():
    print('Evaluated!')
    return True
# The second block is skipped because False and X is always False
if False and check():
    pass
else:
    print('Short-circuited successfully!')
```

### 💻 Example 5: Truthy and Falsy evaluations inside conditions


```python
names = []
if not names:
    print('Empty lists evaluate to Falsy!')
```

### 💻 Example 6: Multiple logical checks combined


```python
age = 25
has_license = True
if age >= 18 and has_license:
    print('Allowed to drive')
```

### 💻 Example 7: Ternary operator inside print


```python
x = 10
print('Even' if x % 2 == 0 else 'Odd')
```

### 💻 Example 8: Guard Clause Pattern (clean alternative to nested if)


```python
def register(age):
    if age < 18:
        return 'Rejected: Underage'
    return 'Registered successfully!'
print(register(16))
print(register(22))
```

### 💻 Example 9: Checking None explicitly in conditions


```python
config = None
if config is None:
    print('No configuration loaded.')
```

## ⚠️ Common Mistakes & Pitfalls
- **Over-nesting conditions:** Creating deeply nested structures that are difficult to read. Use guard clauses instead.
- **Comparing booleans directly:** Writing `if cond == True:` instead of simply `if cond:`.

## 💼 Technical Interview Questions & Answers
**Q1: How does short-circuiting work in Python logical gates?**
*Answer:* For logical `and`, if the first operand is False, Python skips the second operand because the result is guaranteed to be False. For logical `or`, if the first operand is True, the second is skipped.

**Q2: What are falsy values in Python?**
*Answer:* `None`, `False`, numeric zeros (`0`, `0.0`), and empty collections (`''`, `[]`, `()`, `{}`).
