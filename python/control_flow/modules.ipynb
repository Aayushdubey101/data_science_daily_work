## 🔍 Concept Explanation
A module is a file containing Python definitions and statements that can be imported into other scripts.

### 🌟 Why This Matters in Data Science:
Modular designs separate ingestion, transformation, and training pipelines into distinct reusable files.

### 🧠 Core Concepts & Mental Models:
- **Namespace Resolution:** Python executes modules inside their own isolated namespaces, avoiding name collisions.
- **`__name__ == '__main__'`:** Guards executable code so it runs only when the script is executed directly.

---

## 📝 Detailed Explanations
When a module is imported, Python searches the directories listed in `sys.path` (including the current directory, environment paths, and standard libraries).



### 💻 Example 1: Standard library import and execution


```python
import math
print('Square Root of 25:', math.sqrt(25))
print('Value of Pi:', math.pi)
```

### 💻 Example 2: Importing specific objects using 'from'


```python
from math import sin, pi
print('Sin pi/2:', sin(pi / 2))
```

### 💻 Example 3: Namespace aliasing using 'as'


```python
import sys as system
print('Platform:', system.platform)
```

### 💻 Example 4: Listing module exports programmatically with dir()


```python
import math
print('Math exports count:', len(dir(math)))
print('First 5 exports:', dir(math)[:5])
```

### 💻 Example 5: Inspecting script execution scope variable __name__


```python
print('Current Module Namespace __name__:', __name__)
```

### 💻 Example 6: Importing standard libraries: sys module


```python
import sys
print('Executable location:', sys.executable != '')
```

### 💻 Example 7: Importing standard libraries: os module


```python
import os
print('Current working directory:', os.getcwd() != '')
```

## ⚠️ Common Mistakes & Pitfalls
- **Circular imports:** Creating two modules that import each other, causing execution locks.
- **Shadowing standard modules:** Naming a local script `math.py` or `random.py`, which blocks standard library imports.

---

## 💼 Technical Interview Questions & Answers
**Q1: What is the purpose of if __name__ == '__main__':?**
*Answer:* It acts as a guard that allows code to run only when the script is executed directly, and not when it is imported as a module.

**Q2: How does Python locate imported modules?**
*Answer:* It searches the directories listed in the `sys.path` list.
