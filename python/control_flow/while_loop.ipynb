## 🔍 Concept Explanation
A `while` loop repeatedly executes code as long as its condition remains True.

### 🌟 Why This Matters in Data Science:
Gradient descents and continuous pipelines rely on loop convergence checks that execute until convergence criteria are met.

### 🧠 Core Concepts & Mental Models:
- **Sentinel values:** Control variables that trigger loop termination.
- **while-else:** Executes only when the loop terminates normally without hitting a break.

---

## 📝 Detailed Explanations
While loops evaluate their test expressions before every iteration. If the expression evaluates to False initially, the loop body is never executed.



### 💻 Example 1: Sentinel-controlled while loop


```python
count = 0
while count < 3:
    print('Count:', count)
    count += 1
```

### 💻 Example 2: Aborting while loop using break


```python
i = 1
while i < 10:
    if i == 4:
        break
    print('i:', i)
    i += 1
```

### 💻 Example 3: Skipping iterations using continue


```python
i = 0
while i < 4:
    i += 1
    if i == 2:
        continue
    print('Iteration:', i)
```

### 💻 Example 4: The optional while-else construct (runs to natural completion)


```python
i = 1
while i <= 3:
    print('Run:', i)
    i += 1
else:
    print('Else statement executed!')
```

### 💻 Example 5: While-else skipped due to break statement


```python
i = 1
while i <= 3:
    if i == 2:
        print('Aborting loop via break')
        break
    i += 1
else:
    print('Else block executed!')
```

### 💻 Example 6: Simulating a standard DO-WHILE loop structure


```python
count = 1
while True:
    print('Do task:', count)
    count += 1
    if count > 2:
        break
```

### 💻 Example 7: Flag-controlled execution loop


```python
is_running = True
step = 0
while is_running:
    step += 1
    if step >= 2:
        is_running = False
print('Loop exited, final step:', step)
```

### 💻 Example 8: Removing elements from a list using while


```python
fruits = ['apple', 'banana']
while fruits:
    print('Removed:', fruits.pop())
```

## ⚠️ Common Mistakes & Pitfalls
- **Infinite loop crashes:** Forgetting to increment or update the sentinel control variable, causing infinite execution.
- **Misunderstanding while-else:** Thinking the else block runs if the loop is aborted via break (it does not).

---

## 💼 Technical Interview Questions & Answers
**Q1: Under what conditions does the else block of a while loop execute?**
*Answer:* It executes only when the while loop condition evaluates to False naturally, and not when the loop is terminated by a break statement.

**Q2: What is an infinite loop and how do you prevent it?**
*Answer:* A loop that never terminates because its condition is always True. Prevent this by ensuring the control variable is correctly updated in every iteration.
