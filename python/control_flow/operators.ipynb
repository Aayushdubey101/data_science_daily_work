## 🔍 Concept Explanation
Python features arithmetic, comparison, logical, bitwise, identity, and membership operators.

### 🌟 Why This Matters in Data Science:
Data filtering (Pandas masks) and numerical computing rely heavily on operator evaluations.

### 🧠 Core Concepts & Mental Models:
- **== vs is:** `==` checks value equality; `is` checks memory reference/address.
- **Bitwise vs Logical:** Bitwise evaluates binary bits; Logical evaluates truth value short-circuits.

---

## 📝 Detailed Explanations
Python evaluates operators according to an operator precedence hierarchy (exponentiation first, then multiplication/division, then addition/subtraction).



### 💻 Example 1: Standard Arithmetic Operations


```python
x = 10
y = 3
print('Add:', x + y)
print('Floor Div:', x // y)
print('Modulus:', x % y)
```

### 💻 Example 2: Comparison Operators


```python
a = 5
b = 8
print('a > b:', a > b)
print('a == b:', a == b)
print('a != b:', a != b)
```

### 💻 Example 3: Logical operators


```python
t = True
f = False
print('t and f:', t and f)
print('t or f:', t or f)
print('not t:', not t)
```

### 💻 Example 4: Bitwise operations


```python
a = 5  # 0101
b = 3  # 0011
print('Bitwise AND &:', a & b)
print('Bitwise OR |:', a | b)
print('Bitwise XOR ^:', a ^ b)
```

### 💻 Example 5: Identity operators: is vs == (Value vs Memory Address)


```python
x = [1, 2]
y = [1, 2]
z = x
print('x == y (values same):', x == y)
print('x is y (memory address same):', x is y)
print('x is z (shared memory reference):', x is z)
```

### 💻 Example 6: Membership operators: in and not in


```python
lst = [10, 20, 30]
print('20 in lst:', 20 in lst)
print('40 not in lst:', 40 not in lst)
```

### 💻 Example 7: Operator Precedence Demonstration


```python
result = 10 + 2 * 5 ** 2
print('Precedence result:', result)
```

### 💻 Example 8: Multiple assignments with arithmetic


```python
x = 10
x += 5
print('Augmented Add:', x)
```

### 💻 Example 9: Divmod helper function


```python
q, r = divmod(10, 3)
print('Quotient:', q, 'Remainder:', r)
```

### 💻 Example 10: Shift Bitwise Operators


```python
x = 4  # 0100
print('Left Shift x << 1:', x << 1)
print('Right Shift x >> 1:', x >> 1)
```

## ⚠️ Common Mistakes & Pitfalls
- **Using 'is' for integers:** Expecting `x is y` to be true for large numbers (caches exist only for -5 to 256).
- **Pandas logical masks:** Forgetting that Pandas requires bitwise operators (`&`, `|`) instead of logical words (`and`, `or`).

## 💼 Technical Interview Questions & Answers
**Q1: Explain the difference between == and is.**
*Answer:* `==` checks whether the values of the operands are equal. `is` checks whether the operands point to the exact same object reference in memory.

**Q2: What is operator precedence in Python?**
*Answer:* The order of evaluation of different operators in an expression, e.g. exponentiation `**` followed by multiplication `*` before addition `+`.
