## 🔍 Concept Explanation
A `for` loop is an iterator-based loop that steps through elements of any iterable sequence.

### 🌟 Why This Matters in Data Science:
Data preprocessing routinely involves looping over collections (rows, lists, tensors).

### 🧠 Core Concepts & Mental Models:
- **Range generator:** Memory-efficient generator producing sequences on-demand.
- **Enumerate:** Built-in helper that yields both indices and values during iteration.

---

## 📝 Detailed Explanations
For loops leverage Python's iteration protocol, calling `__iter__` and `__next__` behind the scenes until a StopIteration exception occurs.



### 💻 Example 1: Iterating over sequence list


```python
fruits = ['apple', 'banana', 'cherry']
for f in fruits:
    print('Fruit:', f)
```

### 💻 Example 2: Using the range() generator function


```python
for i in range(1, 4):
    print('Number:', i)
```

### 💻 Example 3: Tracking indices cleanly with enumerate()


```python
names = ['Aayush', 'Bob']
for idx, val in enumerate(names):
    print(f'Index: {idx} | Name: {val}')
```

### 💻 Example 4: Iterating over string characters


```python
for char in 'Py':
    print('Char:', char)
```

### 💻 Example 5: Nested for loops for matrix layouts


```python
for row in range(2):
    for col in range(2):
        print(f'({row}, {col})', end=' ')
    print()
```

### 💻 Example 6: Optional for-else statement (runs to completion)


```python
for x in range(3):
    print(x)
else:
    print('For-else completed!')
```

### 💻 Example 7: Iterating over dictionary items key-value maps


```python
profile = {'id': 101, 'role': 'Admin'}
for k, v in profile.items():
    print(f'{k} -> {v}')
```

### 💻 Example 8: Range step parameters (skipping steps)


```python
for x in range(0, 10, 3):
    print('Step x:', x)
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Modifying a collection during iteration:** Appending/deleting elements from a list while looping over it, causing skipped indices.
- **Using range(len(lst)):** A non-pythonic style when you only need to iterate over the elements (use `for x in lst` instead).

## 💼 Technical Interview Questions & Answers
**Q1: Why is range() memory-efficient compared to a concrete list?**
*Answer:* `range()` is a generator-like sequence that calculates numbers on-demand rather than allocating all elements in memory simultaneously, requiring O(1) memory.

**Q2: How do you get both index and value in a loop?**
*Answer:* Use the built-in `enumerate(iterable)` function.
