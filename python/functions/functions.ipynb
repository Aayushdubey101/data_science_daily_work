## 🔍 Concept Explanation
A function is a reusable block of code that executes when called.

### 🌟 Why This Matters in Data Science:
Functions are the core building blocks of modular, maintainable, and testable software.

### 🧠 Core Concepts & Mental Models:
- **Parameters vs Arguments:** Parameters are variable names inside definitions; arguments are actual values passed during calls.
- **Docstrings:** Standard PEP 257 string documentation preserved on the `__doc__` metadata attribute.
- **Execution Stack:** Every function call allocates a new frame in memory containing local variable references.

---

## 📝 Detailed Explanations
When a function is called, Python allocates a new frame on the call stack to manage local namespaces. When the function returns, this frame is popped.



### 💻 Example 1: Function Definition and Calling


```python
def say_hello():
    print('Hello!')
say_hello()
```

### 💻 Example 2: Parameters (inputs in def) vs Arguments (values passed in call)


```python
def add(x, y):  # parameters
    return x + y
print('Sum:', add(10, 20))  # arguments
```

### 💻 Example 3: Default and Keyword Arguments


```python
def profile(name, role='User'):
    print(f'{name}: {role}')
profile('Aayush')
profile(role='Admin', name='Bob')
```

### 💻 Example 4: Function Documentation (Docstring)


```python
def get_pi():
    """Returns value of Pi to 2 decimal places."""
    return 3.14
print('Docstring:', get_pi.__doc__)
```

### 💻 Example 5: Function Execution in Memory (inspected dynamically)


```python
import sys
def test_scope():
    frame = sys._getframe()
    print('Current execution frame local variable count:', len(frame.f_locals))
test_scope()
```

## ⚠️ Common Mistakes & Pitfalls
- **Using mutable defaults:** Expecting default lists to reset across calls (they are instantiated once when the function is defined, sharing state across calls).
- **Missing return statements:** Forgetting to return a value, causing the function to return `None` silently.

## 💼 Technical Interview Questions & Answers
**Q1: Explain the difference between parameters and arguments.**
*Answer:* Parameters are variables declared in a function signature. Arguments are actual values passed to these variables when the function is invoked.

**Q2: What is a call stack frame?**
*Answer:* A memory allocation block created dynamically whenever a function is called, storing local variable states and references.
