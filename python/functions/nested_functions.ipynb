## 🔍 Concept Explanation
A nested function is a function defined inside another function, encapsulating helper logic.

### 🌟 Why This Matters in Data Science:
Nested functions encapsulate helper logic and prevent polluting the module-level namespace.

### 🧠 Core Concepts & Mental Models:
- **Scope Encapsulation:** Isolates helper functions inside parent scopes.
- **Accessing parent state:** Inner functions can read variables from their parent's environment.

---

## 📝 Detailed Explanations
A nested function has its own local namespace and has read-only access to its parent function's namespace.



### 💻 Example 1: Encapsulated inner helper function


```python
def outer_calc(x, y):
    def add():
        return x + y
    return add() * 2
print('Calc Result:', outer_calc(5, 10))
```

### 💻 Example 2: Nested function accessing parent variables


```python
def parent(msg):
    def child():
        print('Parent message:', msg)  # read access
    child()
parent('Hello!')
```

### 💻 Example 3: Inner function isolation (cannot be called externally)


```python
def parent():
    def isolated():
        print('Isolated!')
    pass
try:
    isolated()
except NameError as e:
    print('Expected NameError caught:', e)
```

## ⚠️ Common Mistakes & Pitfalls
- **Calling nested functions externally:** Attempting to invoke a nested function outside its parent's namespace, raising a `NameError`.
- **Deep nesting:** Creating excessively nested functions, making code difficult to read and maintain.

---

## 💼 Technical Interview Questions & Answers
**Q1: Can a nested function be called from outside its parent function?**
*Answer:* No, it is local to the parent function and cannot be accessed from outside.

**Q2: What is scope encapsulation?**
*Answer:* The practice of hiding helper routines inside a function's local scope so they are not exposed externally.
