## 🔍 Concept Explanation
Variable scope determines where a variable can be accessed or modified in code.

### 🌟 Why This Matters in Data Science:
Scoping errors lead to unintended mutations, variable namespace clutter, and logical bugs.

### 🧠 Core Concepts & Mental Models:
- **LEGB Rule:** The resolution search path: Local, Enclosing, Global, Built-in.
- **global / nonlocal:** Keywords that declare a variable refers to an outer scope rather than instantiating a new local variable.

---

## 📝 Detailed Explanations
Python resolves variable names using the LEGB rule. By default, you can read outer-scope variables but cannot modify them; attempting to do so creates a new local variable unless declared `global` or `nonlocal`.



### 💻 Example 1: LEGB scope resolution demonstration


```python
x = 'global'  # Global
def outer():
    x = 'enclosing'  # Enclosing
    def inner():
        x = 'local'  # Local
        print('Inner x:', x)
    inner()
outer()
```

### 💻 Example 2: Modifying global variables using the global keyword


```python
count = 0
def increment():
    global count
    count += 1
increment()
print('Global count:', count)
```

### 💻 Example 3: Modifying enclosing scope variables using nonlocal


```python
def outer():
    value = 'orig'
    def inner():
        nonlocal value
        value = 'new'
    inner()
    print('Outer value after nonlocal:', value)
outer()
```

### 💻 Example 4: Variable shadowing (local hides global)


```python
x = 100
def func():
    x = 200  # Shadows global x
    print('Local x:', x)
func()
print('Global x remains:', x)
```

## ⚠️ Common Mistakes & Pitfalls
- **Excessive use of global:** Using the `global` keyword extensively, which makes code difficult to test and maintain.
- **Confusing global and nonlocal:** Using `global` inside nested functions when `nonlocal` is required to modify enclosing scopes.

---

## 💼 Technical Interview Questions & Answers
**Q1: What is the LEGB rule?**
*Answer:* It is the order Python searches to resolve variable names: Local scope, then Enclosing (outer function) scope, then Global (module level) scope, and finally the Built-in scope.

**Q2: When is the nonlocal keyword used?**
*Answer:* It is used inside nested functions to declare that a variable refers to a binding in the nearest enclosing outer scope, excluding the global scope.
