## 🔍 Concept Explanation
`map(func, iterable)` applies a function to all items in an input list. `filter(func, iterable)` filters elements based on a boolean criteria. `reduce(func, iterable)` cumulatively applies a function to reduce a sequence to a single value.

### 🌟 Why This Matters in Data Science:
Map, filter, and reduce operations enable declarative, memory-efficient data processing pipelines.

### 🧠 Core Concepts & Mental Models:
- **Lazy Evaluation:** `map` and `filter` return lazy iterators that calculate values on-demand.
- **Folds:** `reduce` cumulatively folds sequences into a single aggregated value.

---

## 📝 Detailed Explanations
`map` and `filter` return iterator objects, generating values only when iterated over, keeping memory usage constant O(1).



### 💻 Example 1: Vectorized mapping using map()


```python
nums = [1, 2, 3]
squared = map(lambda x: x**2, nums)
print('Map Iterator:', squared)
print('List mapping:', list(squared))
```

### 💻 Example 2: Conditional filtering using filter()


```python
nums = [1, 2, 3, 4]
evens = filter(lambda x: x % 2 == 0, nums)
print('Filter Iterator:', evens)
print('Filtered list:', list(evens))
```

### 💻 Example 3: Aggregation folding using functools.reduce()


```python
from functools import reduce
nums = [1, 2, 3, 4]
total_sum = reduce(lambda x, y: x + y, nums)
print('Reduced sum:', total_sum)
```

### 💻 Example 4: Combining map and filter operations


```python
nums = [1, 2, 3, 4]
squared_evens = map(lambda x: x**2, filter(lambda x: x % 2 == 0, nums))
print('Squared Evens:', list(squared_evens))
```

## ⚠️ Common Mistakes & Pitfalls
- **Converting iterators to lists repeatedly:** Casting results to lists immediately, losing the memory benefits of lazy evaluation.
- **Using reduce where sum() suffices:** Overcomplicating basic operations with `reduce` when simpler built-in functions exist.

---

## 💼 Technical Interview Questions & Answers
**Q1: Why do map() and filter() return iterator objects instead of concrete lists?**
*Answer:* To enable lazy evaluation, saving memory by generating elements on-demand rather than allocating a full list in memory.

**Q2: Which module must be imported to use the reduce() function?**
*Answer:* The standard library `functools` module.
