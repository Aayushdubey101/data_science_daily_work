## 🔍 Concept Explanation
A higher-order function is a function that either accepts another function as an argument, returns a function, or both.

### 🌟 Why This Matters in Data Science:
Higher-order functions enable declarative programming patterns, allowing you to pass functional logic directly into processing pipelines.

### 🧠 Core Concepts & Mental Models:
- **First-Class Functions:** In Python, functions are objects, allowing assignment, deletion, and parameter nesting.
- **Function Assignment & Deletion:** Assigning function objects to new names and deleting original name bindings.

---

## 📝 Detailed Explanations
Because functions are objects, deleting a function name with `del` only removes the specific binding. The function object remains in memory as long as other references exist.



### 💻 Example 1: Function Assignment (functions as first-class objects)


```python
def greet(name):
    return f'Hello {name}'
say = greet  # assignment
print(say('Aayush'))
```

### 💻 Example 2: Function as an Argument


```python
def run_operation(func, x):
    return func(x)
square = lambda val: val**2
print('Result:', run_operation(square, 5))
```

### 💻 Example 3: Function as a Return Value


```python
def get_multiplier(n):
    def multiplier(val):
        return val * n
    return multiplier
double = get_multiplier(2)
print('Double 10:', double(10))
```

### 💻 Example 4: Function Deletion (deleting references)


```python
def test_func():
    return 'Running'
test_ref = test_func
del test_func  # original name deleted
try:
    test_func()
except NameError as e:
    print('test_func NameError:', e)
    print('Assigned ref still works:', test_ref())
```

## ⚠️ Common Mistakes & Pitfalls
- **Passing function calls instead of references:** Passing `func()` instead of `func` as an argument, executing the function prematurely.
- **Forgetting to invoke returned functions:** Calling the parent function and expecting the inner function to run automatically.

## 💼 Technical Interview Questions & Answers
**Q1: What does it mean that functions are 'first-class citizens' in Python?**
*Answer:* It means functions can be treated like any other object, such as being assigned to variables, passed as arguments, or returned from other functions.

**Q2: What happens when you delete a function name using 'del'?**
*Answer:* The specific variable name reference is removed from the namespace. The underlying function object remains in memory as long as other variables still point to it.
