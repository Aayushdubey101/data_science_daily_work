## 🔍 Concept Explanation
Lambda functions are single-line, anonymous functions defined without a name.

### 🌟 Why This Matters in Data Science:
Lambdas provide a clean, inline way to define quick, throwaway helper functions.

### 🧠 Core Concepts & Mental Models:
- **Anonymous syntax:** Defined using the `lambda` keyword: `lambda args: expression`.
- **Single-expression limit:** Lambdas can only evaluate and return a single expression.

---

## 📝 Detailed Explanations
Lambda functions are syntactically restricted compared to standard functions; they cannot contain statements like `assert`, `pass`, or `return` (returns are implicit).



### 💻 Example 1: Standard anonymous lambda evaluation


```python
square = lambda x: x**2
print('Square of 5:', square(5))
```

### 💻 Example 2: Lambda function with multiple parameters


```python
add = lambda x, y: x + y
print('Sum:', add(10, 20))
```

### 💻 Example 3: Sorting a list of tuples using a lambda key


```python
pairs = [(1, 'one'), (3, 'three'), (2, 'two')]
pairs.sort(key=lambda item: item[0])
print('Sorted pairs:', pairs)
```

### 💻 Example 4: Conditional expression inside a lambda


```python
check = lambda x: 'Even' if x % 2 == 0 else 'Odd'
print('10 is:', check(10))
print('7 is:', check(7))
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Complex lambdas:** Writing complex, multi-statement logic inside lambdas (use standard `def` blocks instead).
- **Named lambdas:** Assigning lambdas to variables (use `def` for named functions).

---

## 💼 Technical Interview Questions & Answers
**Q1: What is a lambda function in Python?**
*Answer:* An anonymous, single-line function defined using the `lambda` keyword, which evaluates and returns a single expression.

**Q2: How do you sort a list of dictionaries by a specific key using a lambda?**
*Answer:* Pass `key=lambda x: x['key_name']` to the `sort()` or `sorted()` function.
