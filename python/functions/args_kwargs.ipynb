## 🔍 Concept Explanation
`*args` and `**kwargs` allow a function to accept a variable number of arguments dynamically.

### 🌟 Why This Matters in Data Science:
Designing wrappers, logging hooks, and decorator APIs requires forwarding arguments seamlessly.

### 🧠 Core Concepts & Mental Models:
- **`*args`:** Unpacks positional arguments into a tuple.
- **`**kwargs`:** Unpacks keyword arguments into a dictionary.

---

## 📝 Detailed Explanations
When a function defines `*args` and `**kwargs`, Python aggregates unmatched positional arguments into a tuple, and unmatched keyword arguments into a dictionary.



### 💻 Example 1: Variable-length positional arguments *args


```python
def sum_all(*args):
    print('args tuple:', args)
    return sum(args)
print('Sum:', sum_all(10, 20, 30))
```

### 💻 Example 2: Variable-length keyword arguments **kwargs


```python
def display_profile(**kwargs):
    print('kwargs dict:', kwargs)
    for k, v in kwargs.items():
        print(f'{k} -> {v}')
display_profile(name='Aayush', role='Mentor')
```

### 💻 Example 3: Combining positional arguments, *args, and **kwargs


```python
def complex_function(a, *args, **kwargs):
    print('a:', a)
    print('args:', args)
    print('kwargs:', kwargs)
complex_function(1, 2, 3, x=10, y=20)
```

### 💻 Example 4: Argument unpacking using * and ** during function calls


```python
def greet(name, role):
    print(f'{name} is a {role}')
# Unpacking dict
profile = {'name': 'Aayush', 'role': 'Mentor'}
greet(**profile)
```

## ⚠️ Common Mistakes & Pitfalls
- **Wrong parameter order:** Declaring `**kwargs` before `*args` in function signatures (the correct order is: positional, *args, keyword-only, **kwargs).
- **Missing unpacking operators:** Forgetting the `*` or `**` operators when forwarding arguments, passing raw tuples/dicts instead.

---

## 💼 Technical Interview Questions & Answers
**Q1: What datatypes do *args and **kwargs resolve to within a function?**
*Answer:* `*args` resolves to a tuple, and `**kwargs` resolves to a dictionary.

**Q2: What is the correct order of parameters in a function signature?**
*Answer:* Standard positional arguments, then `*args`, then keyword-only arguments, and finally `**kwargs`.
