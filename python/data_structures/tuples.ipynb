## 🔍 Concept Explanation
Tuples are immutable sequences, meaning their elements cannot be changed after creation.

### 🌟 Why This Matters in Data Science:
Tuples are ideal for returning multiple values from functions, protecting constant settings from mutations, and acting as hashable keys in dictionaries.

### 🧠 Core Concepts & Mental Models:
- **Tuple Packing/Unpacking:** Assigning a sequence to variables dynamically.
- **List vs Tuple:** Lists are mutable and slower; Tuples are immutable, faster, and consume less memory.

---

## 📝 Detailed Explanations
Because tuples are fixed-size and immutable, Python optimizes their memory allocation, allocating a single contiguous block of memory with minimal overhead.



### 💻 Example 1: Tuple creation and access


```python
tpl = (10, 20, 30)
print('Element at index 0:', tpl[0])
```

### 💻 Example 2: Tuple packing and unpacking


```python
record = ('Aayush', 22)
name, age = record
print('Name:', name, '| Age:', age)
```

### 💻 Example 3: Tuple Operations: Concatenation and repetition


```python
t1 = (1, 2)
t2 = (3, 4)
print('Concatenation:', t1 + t2)
print('Repetition:', t1 * 2)
```

### 💻 Example 4: Tuple Functions: count() and index()


```python
tpl = (1, 2, 2, 3)
print('Count of 2:', tpl.count(2))
print('Index of 3:', tpl.index(3))
```

### 💻 Example 5: List vs Tuple memory allocations comparison


```python
import sys
print('List footprint (3 items):', sys.getsizeof([1, 2, 3]))
print('Tuple footprint (3 items):', sys.getsizeof((1, 2, 3)))
```

## ⚠️ Common Mistakes & Pitfalls
- **Modifying tuple elements directly:** Attempting to assign elements to a tuple, raising a `TypeError`.
- **Defining single element tuples without a comma:** Declaring `t = (5)` instead of `t = (5,)`.

---

## 💼 Technical Interview Questions & Answers
**Q1: Can a tuple contain mutable elements?**
*Answer:* Yes. The tuple's internal reference values remain immutable, but nested mutable objects (e.g. lists) inside it can be modified in place.

**Q2: Why are tuples faster than lists?**
*Answer:* Tuples are immutable and allocated in a single, fixed block of memory, avoiding the overhead of dynamic sizing.
