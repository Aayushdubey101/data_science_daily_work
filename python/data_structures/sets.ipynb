## 🔍 Concept Explanation
A set is an unordered collection of unique, hashable elements. A frozenset is an immutable set.

### 🌟 Why This Matters in Data Science:
Sets are ideal for removing duplicates, performing mathematical set operations, and executing constant-time membership checks.

### 🧠 Core Concepts & Mental Models:
- **Frozenset:** Truly immutable set collection that can be used as a dictionary key or set element.
- **Set Comprehension:** Concise syntax to create sets using iteration.

---

## 📝 Detailed Explanations
Sets are implemented using hash tables, which allows membership checking (`x in set`) to run in constant time O(1), compared to O(N) linear time for lists.



### 💻 Example 1: Set creation and set functions


```python
st = {1, 2, 3}
st.add(4)
st.discard(2)
print('Modified set:', st)
```

### 💻 Example 2: Set Operations: Union, Intersection, Difference


```python
a = {1, 2}
b = {2, 3}
print('Union:', a | b)
print('Intersection:', a & b)
print('Difference (a - b):', a - b)
```

### 💻 Example 3: Frozenset: Truly immutable set collection


```python
fs = frozenset([1, 2, 3])
try:
    fs.add(4)
except AttributeError as e:
    print('Frozenset error caught:', e)
```

### 💻 Example 4: Set Comprehension


```python
unique_evens = {x for x in [1, 2, 2, 3, 4, 4] if x % 2 == 0}
print('Set comprehension:', unique_evens)
```

## ⚠️ Common Mistakes & Pitfalls
- **Creating empty sets with `{}`:** Writing `{}` constructs an empty dictionary. Use `set()` instead.
- **Adding mutable elements:** Attempting to add lists or dictionaries to a set, raising a `TypeError`.

## 💼 Technical Interview Questions & Answers
**Q1: What is a frozenset in Python?**
*Answer:* A frozenset is an immutable, hashable set that cannot be modified after creation, allowing it to be used as a dictionary key.

**Q2: What is the average time complexity of set union operations?**
*Answer:* O(len(a) + len(b)) linear time, as it requires copying elements into a new set.
