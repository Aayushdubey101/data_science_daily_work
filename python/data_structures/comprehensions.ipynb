## 🔍 Concept Explanation
Comprehensions provide a concise syntax to create collections (lists, sets, dictionaries) from existing iterables.

### 🌟 Why This Matters in Data Science:
Comprehensions allow transforming and filtering datasets cleanly in a single, highly optimized line of code.

### 🧠 Core Concepts & Mental Models:
- **Nested Comprehension:** Comprehensions containing multiple nested loops.
- **Syntactic Sugar:** Shorthand syntax replacing standard nested loops.

---

## 📝 Detailed Explanations
Comprehensions are faster than manual loops because they run in an optimized C-level loop inside Python, avoiding the overhead of repeated append calls.



### 💻 Example 1: List Comprehension with standard mapping


```python
squares = [x**2 for x in range(4)]
print('List:', squares)
```

### 💻 Example 2: Set Comprehension


```python
unique_squares = {x**2 for x in [-2, -1, 1, 2]}
print('Set:', unique_squares)
```

### 💻 Example 3: Dictionary Comprehension


```python
power_map = {x: x**2 for x in range(3)}
print('Dict:', power_map)
```

### 💻 Example 4: Nested Comprehension (generating coordinates matrices)


```python
grid = [(x, y) for x in range(2) for y in range(2)]
print('Coordinates Grid:', grid)
```

### 💻 Example 5: Double-nested list flattening


```python
matrix = [[1, 2], [3, 4]]
flattened = [val for row in matrix for val in row]
print('Flattened:', flattened)
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Overly complex nested comprehensions:** Writing multi-level loops inside a single comprehension, severely degrading code readability.
- **Wasting memory:** Using list comprehensions purely for side-effects (like printing) instead of generating values.

---

## 💼 Technical Interview Questions & Answers
**Q1: When should you avoid using list comprehensions?**
*Answer:* When the transformation logic is complex, nested, or requires extensive comments, as it degrades code readability.

**Q2: What is a nested comprehension?**
*Answer:* A comprehension containing multiple nested loops, e.g. `[val for row in matrix for val in row]`.
