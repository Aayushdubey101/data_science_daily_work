## 🔍 Concept Explanation
The `zip(*iterables)` function aggregates elements from each of the iterables, returning an iterator of tuples.

### 🌟 Why This Matters in Data Science:
Mapping columns, combining parallel feature vectors, and building dictionaries from separate lists requires clean alignment of iterables.

### 🧠 Core Concepts & Mental Models:
- **Iterable Alignment:** Automatically groups corresponding elements into tuples.
- **strict=True:** Raises a ValueError if any input iterable has a mismatching length.

---

## 📝 Detailed Explanations
The `zip()` function returns a lazy iterator that aggregates elements on-demand, requiring O(1) memory.



### 💻 Example 1: Zipping two parallel lists


```python
keys = ['name', 'age']
vals = ['Aayush', 22]
paired = list(zip(keys, vals))
print('Paired:', paired)
```

### 💻 Example 2: Constructing dictionaries using zip()


```python
keys = ['name', 'age']
vals = ['Aayush', 22]
user_dict = dict(zip(keys, vals))
print('User Dict:', user_dict)
```

### 💻 Example 3: Zipping unequal length lists (truncates by default)


```python
lst1 = [1, 2, 3]
lst2 = ['a', 'b']
print('Truncated Zip:', list(zip(lst1, lst2)))
```

### 💻 Example 4: Strict zipping in Python 3.10+ (strict=True)


```python
lst1 = [1, 2, 3]
lst2 = ['a', 'b']
try:
    list(zip(lst1, lst2, strict=True))
except ValueError as e:
    print('Expected strict=True ValueError caught:', e)
```

### 💻 Example 5: Unzipping collection tuples using zip(*zip_obj)


```python
paired = [('a', 1), ('b', 2)]
keys, vals = zip(*paired)
print('Keys:', keys, 'Vals:', vals)
```

### 💻 Example 6: Zipping three parallel sequences


```python
names = ['Aayush', 'Bob']
marks = [90, 85]
rolls = [1, 2]
for name, mark, roll in zip(names, marks, rolls):
    print(f'Roll {roll}: {name} got {mark}')
```

## ⚠️ Common Mistakes & Pitfalls
- **Discarding elements silently:** Zipping unequal length lists without realizing that elements from the longer list are silently discarded.
- **Forgetting strict checks:** Omitting `strict=True` when parallel data structures are required to match exactly.

---

## 💼 Technical Interview Questions & Answers
**Q1: What happens when zip() receives iterables of unequal lengths?**
*Answer:* By default, it truncates the output to match the length of the shortest input iterable. In Python 3.10+, you can pass `strict=True` to raise a ValueError instead.

**Q2: How do you unpack zipped elements back into separate sequences?**
*Answer:* Use `zip(*zipped_iterable)`.
