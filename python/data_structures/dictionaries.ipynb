## 🔍 Concept Explanation
Dictionaries are hash-map based associative arrays storing key-value pairs.

### 🌟 Why This Matters in Data Science:
Dictionaries are extensively used to represent JSON payloads, build fast lookups, and aggregate grouped values.

### 🧠 Core Concepts & Mental Models:
- **Nested Dictionary:** Dictionaries containing other dictionaries as values.
- **Dictionary Comprehension:** Shorthand syntax to build dictionaries from iterables.

---

## 📝 Detailed Explanations
Dictionaries allocate hash tables. When a key is looked up, Python hashes the key, resolves its bucket coordinate, and returns the value in O(1) average time.



### 💻 Example 1: Dictionary creation and get() operations


```python
profile = {'id': 101, 'name': 'Aayush'}
print('Name:', profile.get('name'))
print('Role (default):', profile.get('role', 'User'))
```

### 💻 Example 2: Dictionary Functions: keys(), values(), items()


```python
profile = {'id': 101, 'name': 'Aayush'}
print('Keys:', list(profile.keys()))
print('Values:', list(profile.values()))
```

### 💻 Example 3: Nested Dictionaries representation


```python
database = {
    'user_1': {'name': 'Aayush', 'role': 'Admin'},
    'user_2': {'name': 'Bob', 'role': 'User'}
}
print('User 1 Role:', database['user_1']['role'])
```

### 💻 Example 4: Dictionary Comprehension


```python
cubes_map = {x: x**3 for x in range(1, 4)}
print('Cubes map:', cubes_map)
```

## ⚠️ Common Mistakes & Pitfalls
- **Accessing missing keys directly:** Writing `dict[key]` instead of `dict.get(key)`, raising a `KeyError`.
- **Modifying keys during iteration:** Modifying dictionary keys while looping over the dictionary.

---

## 💼 Technical Interview Questions & Answers
**Q1: How do you access values in a nested dictionary safely?**
*Answer:* Chain safe `.get()` methods: `dict.get('parent', {}).get('child')`.

**Q2: What is the time complexity of retrieving a value from a dictionary?**
*Answer:* O(1) constant time on average, because dictionaries are implemented as hash tables.
