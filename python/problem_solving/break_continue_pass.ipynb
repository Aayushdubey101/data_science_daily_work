## 🔍 Concept Explanation
Modifiers like `break`, `continue`, and `pass` alter control flow within loops and structural blocks.

### 🌟 Why This Matters in Data Science:
Loop modifiers optimize executions (e.g. skipping unnecessary rows, aborting searches early) and act as code stubs.

### 🧠 Core Concepts & Mental Models:
- **break:** Exits the nearest enclosing loop immediately.
- **continue:** Skips the rest of the current iteration, jumping to the next loop pass.
- **pass:** A null statement placeholder that does nothing.

---

## 📝 Detailed Explanations
When a loop hits a break, execution jumps to the first statement outside the loop body.



### 💻 Example 1: Aborting loops immediately using break


```python
for i in range(5):
    if i == 3:
        break
    print('i:', i)
```

### 💻 Example 2: Skipping loop iterations using continue


```python
for i in range(5):
    if i == 2:
        continue
    print('i:', i)
```

### 💻 Example 3: Structural pass placeholders


```python
def complex_function():
    pass  # syntactical placeholder
print('Function parsed successfully using pass')
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Using pass instead of continue:** Expecting `pass` to skip loop iterations (it does nothing; use `continue` instead).
- **Diamond loop breaks:** Assuming a break statement inside a nested loop breaks out of the parent loop (it only exits the innermost loop).

---

## 💼 Technical Interview Questions & Answers
**Q1: What is the difference between continue and pass?**
*Answer:* `continue` skips the rest of the current loop iteration and moves to the next one. `pass` is a null statement placeholder that does nothing, letting execution continue normally.

**Q2: Does a break statement in a nested loop break out of the outer loop?**
*Answer:* No, it only terminates the innermost loop.
