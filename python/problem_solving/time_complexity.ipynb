## 🔍 Concept Explanation
Time complexity describes how the execution time of an algorithm scales with the input size (N) using Big-O notation.

### 🌟 Why This Matters in Data Science:
Designing highly scalable code requires identifying bottlenecks, optimizing loops, and managing memory usage.

### 🧠 Core Concepts & Mental Models:
- **Orders of Growth:** O(1) < O(log N) < O(N) < O(N log N) < O(N^2).
- **Cases Analysis:** Best Case (omega), Average Case (theta), and Worst Case (Big-O).

---

## 📝 Detailed Explanations
Nested loop checks iterate recursively. An outer loop running N times with an inner loop running N times results in O(N^2) quadratic time complexity.



### 💻 Example 1: O(1) Constant Time operation


```python
def constant_time(arr):
    return arr[0]  # instant access
print('Result:', constant_time([10, 20, 30]))
```

### 💻 Example 2: O(N) Linear Time operation


```python
def linear_time(arr):
    total = 0
    for x in arr:
        total += x
    return total
print('Sum:', linear_time([1, 2, 3]))
```

### 💻 Example 3: O(N^2) Quadratic Time operation (nested loops)


```python
def quadratic_time(arr):
    pairs = []
    for x in arr:
        for y in arr:
            pairs.append((x, y))
    return len(pairs)
print('Pairs count:', quadratic_time([1, 2, 3]))
```

## ⚠️ Common Mistakes & Pitfalls
- **Assuming built-ins are O(1):** Forgetting that checking list membership (`x in lst`) is an O(N) linear-time search.
- **Ignoring nested method calls inside loops:** Calling O(N) methods (like list slicing or `.index()`) inside loops, turning O(N) loops into O(N^2) complexity.

## 💼 Technical Interview Questions & Answers
**Q1: What is the time complexity of checking if a key exists in a dictionary?**
*Answer:* O(1) constant time on average, because dictionaries are implemented as hash tables.

**Q2: What is the time complexity of searching for an element in a list?**
*Answer:* O(N) linear time, as it requires checking elements sequentially.
