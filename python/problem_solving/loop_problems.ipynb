## 🔍 Concept Explanation
Looping algorithms require careful state management and boundary checks to prevent off-by-one and infinite runtime errors.

### 🌟 Why This Matters in Data Science:
Coding interviews evaluate your ability to manage state variables, handle edge cases, and write optimized search iterations.

### 🧠 Core Concepts & Mental Models:
- **Accumulator variables:** Variables that accumulate values across iterations (e.g., sums, factorials).
- **Loop limits:** Enforcing conditions to prevent infinite loops.

---

## 📝 Detailed Explanations
Optimizing loop steps (e.g. searching divisors up to `int(sqrt(num)) + 1` instead of `num`) reduces time complexity.



### 💻 Example 1: Generating Fibonacci series using loops


```python
n = 5
a, b = 0, 1
fibo = []
for _ in range(n):
    fibo.append(a)
    a, b = b, a + b
print('Fibonacci:', fibo)
```

### 💻 Example 2: Calculating factorial using a loop


```python
num = 5
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print('Factorial of 5:', factorial)
```

### 💻 Example 3: Checking prime numbers with loop optimizations


```python
import math
def is_prime(num):
    if num < 2: return False
    for divisor in range(2, int(math.sqrt(num)) + 1):
        if num % divisor == 0: return False
    return True
print('Is 17 prime?', is_prime(17))
print('Is 20 prime?', is_prime(20))
```

## ⚠️ Common Mistakes & Pitfalls
- **Off-by-one errors:** Boundary mismatches in loop ranges (e.g., `range(1, N)` instead of `range(1, N+1)` for inclusive loops).
- **Failing to handle edge cases:** Ignoring input limits like 0 or negative numbers in sequence generators.

## 💼 Technical Interview Questions & Answers
**Q1: How do you optimize prime checks inside a loop?**
*Answer:* Instead of iterating up to `num-1`, search for divisors only up to `int(sqrt(num)) + 1` to reduce operations from O(N) to O(sqrt(N)).

**Q2: What is an off-by-one error?**
*Answer:* A boundary error where a loop iterates one time too many or one time too few.
