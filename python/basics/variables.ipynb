# 📓 Python Variables & Reference Models
This notebook covers variable declarations, dynamic typing, multiple assignments, memory tracking, and variable swapping.

---

## 🔍 Concept Explanation: Variables as Memory References
In Python, variables are not container boxes; they are **labels or reference pointers** bound to actual objects in memory.

### 🌟 Key Concepts:
1. **Dynamic Typing:** You do not declare variable datatypes. Python determines the type based on the value bound to the label.
2. **Variable Creation:** Variables are initialized the moment a value is assigned to them using the assignment operator `=`.
3. **Reference Identification:** The built-in `id()` function returns the unique integer representing the memory address of the object the variable points to.
4. **Multiple Assignments:** You can bind different labels to values on a single line (e.g. `a = b = c = 10` or `x, y = 5, 10`).
5. **Variable Swapping:** Python allows you to swap variables cleanly using tuple packing/unpacking without a third helper variable: `a, b = b, a`.

## 💻 25 Executable Code Examples
Let's study variable reference allocation:


```python
# Example 1: Variable declaration and assignment
x = 10
print('x:', x)
```

```python
# Example 2: Inspecting dynamic datatype
val = 'Aayush'
print('val type:', type(val))
```

```python
# Example 3: Reassigning a variable to a different type
val = 45.67
print('val type after reassignment:', type(val))
```

```python
# Example 4: Multiple assignment to a single value
a = b = c = 500
print('a:', a, 'b:', b, 'c:', c)
```

```python
# Example 5: Multiple variables pointing to the same memory ID
print('ID of a:', id(a))
print('ID of b:', id(b))
print('Are a and b the same object?', a is b)
```

```python
# Example 6: Reassigning one label shifts its pointer
a = 999
print('Are a and b still same?', a is b)
print('ID of a now:', id(a))
print('ID of b remains:', id(b))
```

```python
# Example 7: Parallel assignment to distinct values
x, y, z = 10, 20.5, 'Python'
print('x:', x, 'y:', y, 'z:', z)
```

```python
# Example 8: Basic variable swapping using a third variable
m = 1
n = 2
temp = m
m = n
n = temp
print('Swapped with temp -> m:', m, 'n:', n)
```

```python
# Example 9: Pythonic variable swapping without a third variable
p = 100
q = 200
p, q = q, p
print('Swapped pythonically -> p:', p, 'q:', q)
```

```python
# Example 10: Swapping variables using arithmetic (integers)
x = 5
y = 8
x = x + y  # x is now 13
y = x - y  # y is now 5
x = x - y  # x is now 8
print('Swapped arithmetically -> x:', x, 'y:', y)
```

```python
# Example 11: Binding variables to complex structures
lst_a = [1, 2, 3]
lst_b = lst_a
print('lst_a ID:', id(lst_a))
print('lst_b ID:', id(lst_b))
print('Pointing to same?', lst_a is lst_b)
```

```python
# Example 12: Mutating mutable objects modifies shared references
lst_a.append(4)
print('Modified lst_a, checking lst_b:', lst_b)
```

```python
# Example 13: Variable names case sensitivity
age = 20
Age = 30
AGE = 40
print(age, Age, AGE)
```

```python
# Example 14: Declaring valid variable names (PEP 8 snake_case)
total_user_count = 1500
print('snake_case variable:', total_user_count)
```

```python
# Example 15: Single underscore variable (often represents throwaway values in loops)
for _ in range(2):
    print('Loop run')
```

```python
# Example 16: Constant conventions in uppercase (still reassignable, but signals read-only to developers)
PI_CONSTANT = 3.14159
print('Pi constant:', PI_CONSTANT)
```

```python
# Example 17: Variables reference to standard library modules
import math
my_sqrt = math.sqrt
print('Using variable reference to calculate sqrt of 9:', my_sqrt(9))
```

```python
# Example 18: Reassigning built-in function names (shadowing - AVOID this in real code!)
# We shadow 'sum' locally and then clean it up
original_sum = sum
sum = 100
print('Shadowed sum:', sum)
# Restoring to avoid breaking downstream notebooks
sum = original_sum
```

```python
# Example 19: Inspecting the integer object caching mechanism (small integers -5 to 256 are cached)
x = 250
y = 250
print('Are x and y caching matched?', x is y)
```

```python
# Example 20: Caching mechanism limits (larger integers are not cached, creating separate objects)
x = 500
y = 500
print('Are large ints x and y caching matched?', x is y)
```

```python
# Example 21: Garbage collection trigger simulation
a = [1, 2, 3] # list created, reference count = 1
a = None      # reference count = 0, original list is swept by garbage collector
print('Variable a reassigned:', a)
```

```python
# Example 22: Variable assignments using arithmetic calculations
x = 10
x += 5 # x = x + 5
print('Augmented assignment x:', x)
```

```python
# Example 23: Assigning function return values to variables
val_len = len('Aayush')
print('val_len:', val_len)
```

```python
# Example 24: Unpacking list values to variables
coords = [10.5, 20.5]
lat, lon = coords
print('Latitude:', lat, 'Longitude:', lon)
```

```python
# Example 25: Unpacking variables with wildcard catch-all '*'
first, *rest = [1, 2, 3, 4]
print('First:', first, 'Rest:', rest)
```

## 💼 Technical Interview Questions & Answers
**Q1: Explain variable declaration and reference binding in Python.**
*Answer:* In languages like C, a variable is a named memory location that stores a value. In Python, variables are references or labels that point to objects in memory. The object exists independently, and assigning a variable merely binds a name/label to that object.

**Q2: What is dynamic typing?**
*Answer:* Dynamic typing means variables can be bound to objects of different datatypes over their lifetime without explicit declaration. The interpreter determines the datatype dynamically based on the value currently assigned.

**Q3: What does the built-in id() function return?**
*Answer:* It returns the unique integer memory address of the object the variable points to. If two variables have the same `id()`, they point to the exact same object in memory.

**Q4: Explain Python's small integer caching.**
*Answer:* Python pre-allocates and caches integers in the range `-5` to `256` in memory. When you assign any variable in this range, it points to the cached object instead of creating a new one, meaning `is` returns `True`. For values outside this range, new objects are constructed.
