## 🔍 Concept Explanation: The Anatomy of print()
The built-in `print()` function writes outputs to standard output (usually your screen).

```python
print(*objects, sep=' ', end='\n', file=sys.stdout, flush=False)
```

### 🌟 Key Parameters to Master:
1. **`*objects`:** Any number of positional arguments to print.
2. **`sep`:** The separator string placed between items. Defaults to a single space `' '`.
3. **`end`:** The trailing string appended at the end of the printed values. Defaults to a newline `'\n'`.
4. **Escape Sequences:** Characters like `\n` (newline), `\t` (tab), and `\\` (literal backslash).
5. **Formatting Options:** From old-style percentage `%` operators to modern **f-strings** (formatted string literals).

---

## 💻 25 Executable Code Examples
Let's see these parameters and formats in action:


````python
# Example 1: Standard string print
print('Hello World')
````

````python
# Example 2: Printing multiple items separated by commas (default is space)
print('Aayush', 22, 'Data Scientist')
````

````python
# Example 3: Custom separator parameter 'sep'
print('Aayush', 22, 'Data Science', sep=' | ')
````

````python
# Example 4: Creating a comma-separated row (CSV style)
print('apple', 'banana', 'cherry', sep=',')
````

````python
# Example 5: Dense packing with sep empty string
print('10', '02', '2026', sep='')
````

````python
# Example 6: Custom end parameter (printing on the same line)
print('Hello', end=' ')
print('World!')
````

````python
# Example 7: Connecting outputs with arrows using end
print('Load Data', end=' -> ')
print('Process Data', end=' -> ')
print('Save Model')
````

````python
# Example 8: Multi-parameter combinations (both sep and end customized)
print('CPU', 'GPU', sep='+', end=' = POWER\n')
print('Done')
````

````python
# Example 9: Newline escape sequence \n
print('Line1\nLine2\nLine3')
````

````python
# Example 10: Tab alignment escape sequence \t
print('Name:\tAayush\nAge:\t22')
````

````python
# Example 11: Escaping backslashes for path directory formats
print('C:\\Users\\daayu\\Downloads')
````

````python
# Example 12: Escaping single quotes inside a single-quoted string
print('It\'s a fantastic day for programming.')
````

````python
# Example 13: Carriage return \r (replaces characters from start of line)
print('Hello World\rPython')
````

````python
# Example 14: Old-style %-formatting string interpolation
name = 'Aayush'
print('Hello, %s!' % name)
````

````python
# Example 15: Old %-formatting with integers
age = 22
print('Age: %d years old.' % age)
````

````python
# Example 16: Floating-point precision styling using %
pi_val = 3.14159
print('Pi: %.2f' % pi_val)
````

````python
# Example 17: Standard bracket alignment with .format()
print('The {} is a great language for {}'.format('Python', 'Data Science'))
````

````python
# Example 18: Positional indexing inside .format()
print('Coordinate: ({0}, {1}, {0})'.format(5, 10))
````

````python
# Example 19: Named keyword parameters inside .format()
print('User: {usr}, Role: {role}'.format(usr='Aayush', role='Mentor'))
````

````python
# Example 20: Modern f-string interpolation
username = 'Aayush'
print(f'Logged in as: {username}')
````

````python
# Example 21: Evaluation of mathematical expressions inside f-strings
print(f'Sum: 10 + 20 = {10 + 20}')
````

````python
# Example 22: Call class methods inside f-strings dynamically
title = 'data science'
print(f'Standard Title: {title.title()}')
````

````python
# Example 23: Float precision formatting inside f-strings
accuracy = 0.987654
print(f'Model Accuracy: {accuracy:.4f}')
````

````python
# Example 24: Zero padding numbers inside f-strings
epoch_num = 5
print(f'Epoch: {epoch_num:03d}')
````

````python
# Example 25: Column width formatting inside f-strings
item = 'Apple'
price = 1.25
print(f'{item:<10} : ${price:>5.2f}')
````

## 💼 Technical Interview Questions & Answers
**Q1: What does the end parameter do in print(), and what is its default value?**
*Answer:* The `end` parameter defines the character appended at the end of the printed sequence. Its default value is the newline character `'\n'`.

**Q2: How can you output backslashes in path strings without triggering escape sequences?**
*Answer:* You can either double-escape the backslashes (`\\`) or prefix the string literal with an `r` to make it a raw string (e.g. `r'C:\temp'`).

**Q3: How do you format a float variable 'val' to 2 decimal places using an f-string?**
*Answer:* Use the precision formatting token: `f'{val:.2f}'`.

**Q4: Can print() write to file streams directly instead of the terminal console?**
*Answer:* Yes. By default, `file=sys.stdout`. You can direct it to an open file stream using `print(..., file=open_file_object)`.
