## 🔍 Concept Explanation: Types of Literals
A literal is raw data passed directly into a variable or expression. Python supports several literal types:

### 🌟 Core Literal Classifications:
1. **Numeric Literals:**
    *   **Integer Literals:** Binary (prefix `0b`), Octal (prefix `0o`), Hexadecimal (prefix `0x`), or standard Decimal.
    *   **Float Literals:** Standard decimal decimals (e.g. `3.14`) or exponential notation (`1.2e3`).
    *   **Complex Literals:** Written with a trailing `j` (e.g., `3 + 4j`).
2. **String Literals:** Single, double, or triple-quoted text blocks. Triple quotes allow multi-line strings.
3. **Boolean Literals:** `True` and `False`.
4. **Special Literals:** `None` (representing null states).
5. **Collection Literals:** Direct declarations of lists `[...]`, tuples `(...)`, sets `{...}`, and dicts `{k: v}`.
6. **Raw Strings:** Strings prefixed with `r` that ignore escape sequences (e.g. treating `\n` as a literal backslash and 'n').

## 💻 20 Executable Code Examples
Let's inspect literal representation:


```python
# Example 1: Standard decimal integer literal
dec_lit = 15
print('Decimal:', dec_lit)
```

```python
# Example 2: Binary integer literal (prefix 0b)
bin_lit = 0b1111
print('Binary 0b1111 value in decimal:', bin_lit)
```

```python
# Example 3: Octal integer literal (prefix 0o)
oct_lit = 0o17
print('Octal 0o17 value in decimal:', oct_lit)
```

```python
# Example 4: Hexadecimal integer literal (prefix 0x)
hex_lit = 0xF
print('Hexadecimal 0xF value in decimal:', hex_lit)
```

```python
# Example 5: Float literal decimal notation
flo_lit = 3.14159
print('Float Literal:', flo_lit)
```

```python
# Example 6: Float literal scientific exponential notation
exp_lit = 2.5e3
print('Exponential 2.5e3 representation:', exp_lit)
```

```python
# Example 7: Complex literal (imaginary j)
cplx_lit = 2 + 5j
print('Complex Literal:', cplx_lit)
```

```python
# Example 8: Double-quoted vs single-quoted string literal
str1 = 'Hello'
str2 = "World"
print(str1, str2)
```

```python
# Example 9: Multi-line string literal using triple quotes
multiline_lit = """Line 1
Line 2"""
print('Multi-line String:\n', multiline_lit, sep='')
```

```python
# Example 10: Unicode escape string literal representation
unicode_char = '\u2705'
print('Unicode Character Check Mark:', unicode_char)
```

```python
# Example 11: Escaped string literal pathing
escaped_path = 'C:\\Users\\daayu\\Documents'
print('Escaped path string:', escaped_path)
```

```python
# Example 12: Raw string literal representation (prefix r)
raw_path = r'C:\Users\daayu\Documents'
print('Raw path string:', raw_path)
```

```python
# Example 13: Special Literal None representation
status = None
print('None Special Literal:', status)
```

```python
# Example 14: List collection literal declaration
lst_lit = [1, 'two', 3.0]
print('List Literal:', lst_lit)
```

```python
# Example 15: Tuple collection literal declaration
tpl_lit = (10, 20)
print('Tuple Literal:', tpl_lit)
```

```python
# Example 16: Set collection literal declaration
set_lit = {1, 2, 3}
print('Set Literal:', set_lit)
```

```python
# Example 17: Dictionary collection literal declaration
dict_lit = {'a': 1, 'b': 2}
print('Dictionary Literal:', dict_lit)
```

```python
# Example 18: Boolean truth literal states
truth_lit = True
print('Boolean True:', truth_lit)
```

```python
# Example 19: Float literal with negative exponent
neg_exp = 3.5e-2
print('Negative Exponent 3.5e-2 representation:', neg_exp)
```

```python
# Example 20: Character literal (single length string)
char_lit = 'A'
print('Character:', char_lit, '| Length:', len(char_lit))
```

## 💼 Technical Interview Questions & Answers
**Q1: What are raw strings in Python, and when are they useful?**
*Answer:* A raw string is defined by prefixing a string literal with `r` or `R` (e.g., `r'string'`). It instructs Python to ignore all backslash escape sequences (like `\n` or `\t`), treating backslashes as literal characters. This is highly useful when writing Windows file paths or regex patterns.

**Q2: What is the decimal equivalent of the hexadecimal literal 0x1A?**
*Answer:* 0x1A equals 26 in decimal (`1 * 16^1 + 10 * 16^0`).

**Q3: What does the special literal None represent?**
*Answer:* `None` represents a null state, the absence of a value, or a default return state. It is an instance of the class `NoneType` and evaluates to `False` in boolean contexts.

**Q4: How do you declare a single-element tuple literal in Python?**
*Answer:* You must include a trailing comma (e.g. `t = (5,)`). Writing `t = (5)` simply places the integer 5 inside parentheses, which evaluates to a standard integer of type `int`.
