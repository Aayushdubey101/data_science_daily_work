# 📓 Handling User Input & Defensive Validation
This notebook covers capturing input streams using the `input()` function, type coercion, and building mini-application forms with validation.

---

## 🔍 Concept Explanation: Capturing User Input
The `input()` function suspends program execution and waits for standard input from the user.

### 🌟 Key Rules:
1. **String Default:** `input()` **always returns a string (`str`)** regardless of what the user types.
2. **Type Coercion:** If numerical values are expected, the input must be explicitly cast using `int()` or `float()`.
3. **Defensive Validation:** Wrap numerical conversions inside `try-except` blocks catching `ValueError` to handle invalid inputs cleanly without crashing.

## 🛠️ Mini Application 1: Age Calculator
This script takes a birth year input, converts it to an integer, and calculates the current age based on the year 2026.


```python
# Example 10: Age Calculator
birth_year_input = '2004' # simulated
try:
    birth_year = int(birth_year_input)
    current_year = 2026
    calculated_age = current_year - birth_year
    print(f'Birth Year: {birth_year} | Calculated Age in 2026: {calculated_age} years old.')
except ValueError:
    print('Please enter a valid numeric year.')
```

---

## 🛠️ Mini Application 2: Body Mass Index (BMI) Calculator
This application reads a weight and height, checks for invalid non-numeric inputs, and computes BMI.


```python
# Example 11: BMI Calculator
weight_in = '70' # simulated
height_in = '1.75' # simulated
try:
    weight = float(weight_in)
    height = float(height_in)
    bmi = weight / (height ** 2)
    print(f'Weight: {weight} kg | Height: {height} m | Calculated BMI: {bmi:.2f}')
except ValueError:
    print('Invalid weight or height values provided.')
```

## 🛠️ Mini Application 3: Student Registration Form
Captures multiple attributes (name, email, age, score), performing custom validations on numerical bounds and email formats.


```python
# Example 12: Student Registration Form
name_in = 'Aayush'
email_in = 'aayush@dsmp.com'
age_in = '22'
score_in = '95.5'

try:
    name = name_in.strip()
    email = email_in.strip()
    age = int(age_in)
    score = float(score_in)
    
    if '@' not in email or '.' not in email:
         print('Invalid email format.')
    elif age <= 0:
         print('Age must be a positive integer.')
    else:
         print('[REGISTRATION SUCCESS]')
         print(f'Student: {name} ({age} y/o) | Email: {email} | Exam Score: {score:.1f}%')
except ValueError:
    print('Registration failed: Invalid age or score input.')
```

```python
# Example 14: Checking input truthiness dynamically
simulated_confirmation = 'yes'
is_confirmed = simulated_confirmation.lower().strip() == 'yes'
print('Is Confirmed?', is_confirmed)
```

```python
# Example 15: Handling floats input casting error directly
try:
    int('45.99')
except ValueError as e:
    print('Direct int cast of float string failed:', e)
    print('Corrected float-to-int cast:', int(float('45.99')))
```

## 💼 Technical Interview Questions & Answers
**Q1: What datatype is returned by Python's built-in input() function?**
*Answer:* It always returns a string (type `str`), regardless of whether the user inputs numbers, booleans, or special characters.

**Q2: What happens if you call int() on the string input '3.14'?**
*Answer:* It raises a `ValueError` because the string represents a floating-point value. You must first convert it to a float using `float('3.14')` and then convert it to an integer using `int()`.

**Q3: How do you implement robust input validation in production Python scripts?**
*Answer:* Wrap all conversion operations in a `try-except` structure catching `ValueError`, sanitize input data using `.strip()`, and enforce logical boundaries inside conditional blocks.

**Q4: Explain the difference between compile-time check and runtime evaluation for input variables.**
*Answer:* Python evaluates variables dynamically at runtime. This means inputs must be thoroughly validated at runtime because the interpreter cannot predict what characters a user might enter before the code executes.
