# 📓 Implicit vs Explicit Type Conversion Notes
This notebook covers implicit type promotions, explicit casting constraints, collection conversions, and parsing failures.

---

## 🔍 Concept Explanation: Implicit vs Explicit Conversion
Type conversion is the process of translating a value from one data type to another.

### 🌟 Core Conversions:
1. **Implicit Type Conversion (Automatic Coercion):** Python automatically promotes variables to a wider datatype during operations to prevent data loss (e.g., adding an `int` to a `float` promotes the `int` to a `float`).
2. **Explicit Type Conversion (Casting):** Performed manually by the developer using constructor functions:
    *   `int()`: Converts compatible values to integers (truncating float decimal fractions).
    *   `float()`: Converts numbers or strings to floating-point values.
    *   `str()`: Converts any object into its string representation.
    *   `bool()`: Converts values to boolean state (values representing zero or empty collections resolve to `False`, others resolve to `True`).

## 💻 20 Executable Code Examples
Let's see these conversion behaviors in action:


```python
# Example 1: Implicit conversion - automatic float promotion
x = 10
y = 5.5
sum_val = x + y
print('Value:', sum_val, '| Type:', type(sum_val))
```

```python
# Example 2: Implicit boolean conversion in mathematical operations
b_val = True  # True resolves to 1
result = x + b_val
print('Result of int + bool:', result, '| Type:', type(result))
```

```python
# Example 3: Explicit conversion - string to integer
str_num = '150'
int_num = int(str_num)
print('Casted Integer:', int_num, '| Type:', type(int_num))
```

```python
# Example 4: Explicit conversion - float to integer (decimal truncation)
float_num = 99.99
int_num = int(float_num)
print('Casted from float (truncated):', int_num)
```

```python
# Example 5: Explicit conversion - integer to float
num = 45
flo = float(num)
print('Casted Float:', flo, '| Type:', type(flo))
```

```python
# Example 6: Explicit conversion - string to float
str_flo = '3.14159'
flo = float(str_flo)
print('Casted Float:', flo, '| Type:', type(flo))
```

```python
# Example 7: Explicit conversion - integer to string
count = 100
str_count = str(count)
print('Casted String:', repr(str_count), '| Type:', type(str_count))
```

```python
# Example 8: Casting a list to string representation
lst = [1, 2, 3]
str_lst = str(lst)
print('String representation of list:', repr(str_lst), '| Length:', len(str_lst))
```

```python
# Example 9: Boolean casting - truthy numeric value
print('bool(42):', bool(42))
print('bool(-1):', bool(-1))
```

```python
# Example 10: Boolean casting - falsy numeric value
print('bool(0):', bool(0))
print('bool(0.0):', bool(0.0))
```

```python
# Example 11: Boolean casting - truthy collection (non-empty string)
print('bool("Hello"):', bool('Hello'))
```

```python
# Example 12: Boolean casting - falsy collection (empty string)
print('bool(""):', bool(''))
```

```python
# Example 13: Boolean casting - falsy structures (empty list/dict)
print('bool([]):', bool([]))
print('bool({}):', bool({}))
```

```python
# Example 14: Converting a tuple to a list (enabling modifications)
tpl = (10, 20, 30)
lst = list(tpl)
lst[0] = 99
print('List converted from tuple:', lst)
```

```python
# Example 15: Converting a list to an immutable tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print('Tuple from list:', my_tuple)
```

```python
# Example 16: Converting a list with duplicates to a unique set
dup_list = [1, 2, 2, 3, 3]
unique_set = set(dup_list)
print('Unique Set from list:', unique_set)
```

```python
# Example 17: Converting key-value tuples to a dictionary
pairs = [('name', 'Aayush'), ('role', 'Mentor')]
user_dict = dict(pairs)
print('Dict from tuples list:', user_dict)
```

```python
# Example 18: Conversion failure - non-numeric characters in int()
try:
    int('123a')
except ValueError as e:
    print('Expected Int Conversion failure caught:', e)
```

```python
# Example 19: Conversion failure - float string to int directly
try:
    int('45.67')
except ValueError as e:
    print('Direct float string to int casting failure caught:', e)
```

```python
# Example 20: Complex to float casting failure (explicitly prohibited)
try:
    float(3+4j)
except TypeError as e:
    print('Complex to float TypeError caught:', e)
```

## 💼 Technical Interview Questions & Answers
**Q1: What is the difference between implicit and explicit type conversion?**
*Answer:* Implicit conversion is performed automatically by the interpreter during operations (e.g. promoting an integer to a float when adding them) to prevent data loss. Explicit conversion is performed manually by the developer using constructor functions like `int()`, `float()`, or `str()`.

**Q2: Why does casting float values to integers lose precision?**
*Answer:* Casting to an integer using `int()` does not perform standard rounding. Instead, it performs truncation, discarding the entire fractional/decimal part of the floating-point value (e.g. `int(9.99)` evaluates to `9`).

**Q3: Which values evaluate to False when cast to a boolean using bool()?**
*Answer:* Only falsy values evaluate to `False`. These include numerical zero values (`0`, `0.0`, `0j`), empty collections (`''`, `[]`, `()`, `{}`, `set()`), `None`, and `False` itself. All other values evaluate to `True`.

**Q4: How do you convert a string containing '45.67' into an integer?**
*Answer:* You cannot cast it directly using `int('45.67')`, which raises a `ValueError`. You must first cast the string to a float using `float('45.67')` and then convert that float to an integer using `int()`.
