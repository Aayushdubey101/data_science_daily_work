## 🔍 Concept Explanation: The Data Model
In Python, every value is an object, and every object has a specific **data type**. Python categorizes data types into:

### 🌟 Type Classifications:
*   **Numeric Types:** `int` (integers), `float` (decimals), `complex` (imaginary numbers).
*   **Sequence Types:** `str` (strings), `list` (mutable arrays), `tuple` (immutable sequences).
*   **Boolean Type:** `bool` (logical states: `True`, `False`).
*   **Set Types:** `set` (unordered unique collections).
*   **Mapping Type:** `dict` (associative key-value pairs).

We use the built-in `type()` function to inspect the type classification of any object in memory.

---

## 💻 20 Executable Code Examples
Let's inspect every built-in datatype:


```python
# Example 1: Integer (int) data type
x = 100
print('Value:', x, '| Type:', type(x))
```

```python
# Example 2: Floating point (float) data type
y = 99.95
print('Value:', y, '| Type:', type(y))
```

```python
# Example 3: Floating point scientific notation
scientific_num = 1.2e3
print('Scientific notation:', scientific_num, '| Type:', type(scientific_num))
```

```python
# Example 4: Complex numbers (real + imaginary)
c_num = 3 + 4j
print('Complex value:', c_num, '| Type:', type(c_num))
```

```python
# Example 5: Accessing real and imaginary components
print('Real part:', c_num.real)
print('Imaginary part:', c_num.imag)
```

```python
# Example 6: Boolean (bool) values
is_active = True
print('Boolean value:', is_active, '| Type:', type(is_active))
```

```python
# Example 7: Boolean states from logical statements
evaluation = (10 > 50)
print('Logical output:', evaluation, '| Type:', type(evaluation))
```

```python
# Example 8: String (str) data type
msg = 'Data Science'
print('String value:', msg, '| Type:', type(msg))
```

```python
# Example 9: Multi-line string data type
paragraph = """Line A
Line B"""
print('Multi-line type:', type(paragraph))
```

```python
# Example 10: Mutable Sequence: List (list) data type
lst = [10, 20.5, 'Python']
print('List contents:', lst, '| Type:', type(lst))
```

```python
# Example 11: Modifying a list (demonstrating mutability)
lst[0] = 99
print('Mutated List:', lst)
```

```python
# Example 12: Immutable Sequence: Tuple (tuple) data type
tpl = (10, 20, 30)
print('Tuple contents:', tpl, '| Type:', type(tpl))
```

```python
# Example 13: Unordered Collection of Unique Elements: Set (set) data type
st = {1, 2, 2, 3, 3}
print('Unique set elements:', st, '| Type:', type(st))
```

```python
# Example 14: Adding elements to a set
st.add(4)
print('Updated Set:', st)
```

```python
# Example 15: Key-Value Mapping: Dictionary (dict) data type
user_dict = {'id': 101, 'name': 'Aayush'}
print('Dictionary:', user_dict, '| Type:', type(user_dict))
```

```python
# Example 16: Accessing dictionary values by key
print('Name from dict:', user_dict['name'])
```

```python
# Example 17: Range data type
r = range(5)
print('Range representation:', r, '| Type:', type(r))
```

```python
# Example 18: Checking type of None object
empty_val = None
print('None type:', type(empty_val))
```

```python
# Example 19: Nested structures - list of dictionaries
dataset = [
    {'id': 1, 'score': 90},
    {'id': 2, 'score': 85}
]
print('Nested dataset:', dataset, '| Type:', type(dataset))
```

```python
# Example 20: Immutable set representation - frozenset
fs = frozenset([1, 2, 3])
print('Frozenset:', fs, '| Type:', type(fs))
```

## 💼 Technical Interview Questions & Answers
**Q1: What are the differences between a list and a tuple in Python?**
*Answer:* A list is mutable, meaning you can modify its elements in place after creation, and uses square brackets `[...]`. A tuple is immutable, meaning its elements cannot be changed after creation, and uses parentheses `(...)`. Because of this immutability, tuples are faster and consume less memory.

**Q2: What is an unordered collection in Python, and which data type represents it?**
*Answer:* An unordered collection means elements are not indexed or kept in insertion order. The `set` datatype represents this, allowing fast mathematical unique operations and constant-time membership lookups.

**Q3: Can a dictionary have duplicate keys?**
*Answer:* No. Dictionary keys must be unique. If you assign a value to an existing key, it overwrites the previous value bound to that key.

**Q4: What datatype is None in Python?**
*Answer:* The special value `None` represents the absence of a value or a null state. Its type is `NoneType`.
