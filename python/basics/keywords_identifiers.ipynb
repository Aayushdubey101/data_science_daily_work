# 📓 Python Keywords & Identifiers Reference Notes
This notebook defines identifiers, standard keyword boundaries, identifier rules, naming conventions, and reserved syntax traps.

---

## 🔍 Concept Explanation: Keywords vs Identifiers
*   **Keywords:** Reserved words in Python that have special syntactical meaning. They cannot be used as identifier names (variable names, function names, class names).
*   **Identifiers:** Names given to variables, functions, classes, and objects.

### 🌟 Identifier Naming Rules:
1. Must begin with a letter (A-Z, a-z) or an underscore `_`.
2. Cannot start with a number (e.g. `1_var = 10` is a SyntaxError).
3. Can contain alphanumeric characters and underscores (`A-Z`, `a-z`, `0-9`, `_`).
4. Case-sensitive (`value` is different from `Value`).
5. Cannot be a reserved keyword.

We use the built-in `keyword` module to list all keywords programmatically.

## 💻 15 Executable Code Examples
Let's study naming and keyword verification:


```python
# Example 1: Checking keywords programmatically
import keyword
print('Total keywords:', len(keyword.kwlist))
print('Keywords:', keyword.kwlist[:10])
```

```python
# Example 2: Verifying a keyword using iskeyword()
print('Is "for" a keyword?', keyword.iskeyword('for'))
print('Is "variable" a keyword?', keyword.iskeyword('variable'))
```

```python
# Example 3: Declaring valid variable names (alphanumeric and underscore)
user_id = 101
user1 = 'Aayush'
print(user_id, user1)
```

```python
# Example 4: Case-sensitivity demonstration in identifiers
score = 10
Score = 20
SCORE = 30
print(score, Score, SCORE)
```

```python
# Example 5: Variable beginning with underscore (protected convention)
_internal_limit = 500
print('Internal variable:', _internal_limit)
```

```python
# Example 6: Variable beginning with double underscore (triggers name mangling in classes)
__private_id = 'SECRET'
print('Private ID:', __private_id)
```

```python
# Example 7: Simulating invalid variable name - starting with number
try:
    # We simulate a syntax check using eval to catch the SyntaxError
    eval('1_score = 99')
except SyntaxError as e:
    print('Caught expected SyntaxError:', e)
```

```python
# Example 8: Simulating invalid variable name - containing special characters
try:
    eval('price$ = 100')
except SyntaxError as e:
    print('Special characters syntax error:', e)
```

```python
# Example 9: Simulating invalid variable name - using a reserved keyword
try:
    eval('class = 5')
except SyntaxError as e:
    print('Keyword assignment error:', e)
```

```python
# Example 10: Valid identifier containing numbers in middle
item_4_price = 12.99
print('Valid identifier:', item_4_price)
```

```python
# Example 11: Dynamic checks on identifier validity using str.isidentifier()
print('Is "valid_name" valid?', 'valid_name'.isidentifier())
print('Is "1_invalid" valid?', '1_invalid'.isidentifier())
```

```python
# Example 12: PEP 8 naming conventions - class names (PascalCase)
class UserProfile:
    pass
print('Class identifier defined using PascalCase')
```

```python
# Example 13: PEP 8 naming conventions - function names (snake_case)
def calculate_total_bill(amount):
    return amount * 1.18
print('Function identifier defined using snake_case')
```

```python
# Example 14: Shadowing standard built-ins (valid identifier, but bad practice)
# Shadowing 'list' makes it unavailable as a constructor until cleaned up
original_list = list
list = [10, 20]
print('Shadowed identifier list:', list)
# Restoring standard built-in to keep environment healthy
list = original_list
```

```python
# Example 15: Displaying true keywords (True, False, None are capitalized!)
print('Special capitalized keywords:', keyword.iskeyword('True'), keyword.iskeyword('None'))
```

## 💼 Technical Interview Questions & Answers
**Q1: What is the difference between a keyword and an identifier?**
*Answer:* A keyword is a reserved word defined in Python's core syntax that has special meaning (e.g. `for`, `if`, `class`). An identifier is a user-defined name given to variables, functions, and classes to identify them in code.

**Q2: What are the naming constraints for identifiers in Python?**
*Answer:* Identifiers must start with a letter or an underscore (`_`), can contain alphanumeric characters and underscores, cannot contain special characters like `$`, `@`, or `-`, and cannot be a reserved keyword.

**Q3: Are variable names case-sensitive in Python?**
*Answer:* Yes. Python is fully case-sensitive, meaning `age`, `Age`, and `AGE` are treated as three completely independent identifiers pointing to different objects in memory.

**Q4: Which three keywords in Python start with a capital letter?**
*Answer:* `True`, `False`, and `None`. All other keywords are written in lowercase.
