# 📓 Writing Readable Code: Comments and Docstrings
This notebook details single-line comments, multi-line blocks, docstrings (PEP 257), and formatting best practices.

---

## 🔍 Concept Explanation: Communicating in Code
Comments are ignored by the Python compiler and interpreter. They explain **why** code is written a certain way, rather than **what** the code does.

### 🌟 Key Comment Types:
1. **Single-Line Comments:** Indicated using the hash symbol `#`.
2. **Inline Comments:** Placed on the same line as the statement, separated by at least two spaces.
3. **Multi-Line Comments:** Constructed using consecutive `#` characters on separate lines (or raw strings, though consecutive hashes are PEP 8 standard).
4. **Docstrings (Documentation Strings):** Enclosed inside triple quotes (`"""` or `'''`) as the first statement in a module, class, or function definition. These are preserved at runtime and can be accessed using the `__doc__` metadata attribute or `help()`.

## 💻 15 Executable Code Examples
Let's see comments and docstrings in action:


```python
# Example 1: Basic single-line comment
print('Comments are ignored')
```

```python
x = 10  # Example 2: Inline comment separated by two spaces
print(x)
```

```python
# Example 3: Multi-line comments
# Step 1: Initialize coordinates
# Step 2: Print variables
lat = 12.97
print('Latitude:', lat)
```

```python
# Example 4: Function with standard PEP 257 Docstring
def calculate_area(width, height):
    """Calculates and returns the area of a rectangle."""
    return width * height

print('Area:', calculate_area(5, 10))
```

```python
# Example 5: Accessing __doc__ metadata programmatically
print('Docstring value:', calculate_area.__doc__)
```

```python
# Example 6: multi-line docstring structure with parameters
def compute_bmi(weight_kg, height_m):
    """
    Computes the Body Mass Index.
    
    Parameters:
    weight_kg (float): Weight in kilograms.
    height_m (float): Height in meters.
    
    Returns:
    float: Calculated BMI value.
    """
    return weight_kg / (height_m ** 2)

print('BMI:', round(compute_bmi(70, 1.75), 2))
```

```python
# Example 7: Using help() to read docstrings cleanly
help(calculate_area)
```

```python
# Example 8: Multi-line string literal as a block comment (valid, but not standard PEP 8)
"""
This acts as a multiline comment block.
It is technically a standalone string literal, not a comment.
"""
print('Literal comment parsed successfully')
```

```python
# Example 9: PEP 8 comment spacing rule (at least one space after # symbol)
#Good Comment
# Bad Comment (missing space - non-PEP 8 standard)
print('Spacing rule demo')
```

```python
# Example 10: Temporarily disabling code blocks using comments during debugging
x = 5
# x = 10  # This code line is disabled
print('x is:', x)
```

```python
# Example 11: Docstring on a custom class
class DatabaseConnector:
    """Encapsulates connections to our cloud database repository."""
    pass

print('Class doc:', DatabaseConnector.__doc__)
```

```python
# Example 12: Inline comment styling
price = 100  # Price in USD
tax = 0.05   # 5% state tax
print('Tax sum:', price * tax)
```

```python
# Example 13: Comments detailing complex algorithms (e.g. leap year rules)
year = 2024
# Leap year: divisible by 4, but not by 100 unless also divisible by 400
is_leap = (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0)
print('Is Leap Year?', is_leap)
```

```python
# Example 14: Docstrings inside module imports
import math
print('Math module doc:', math.__doc__.split('.')[0] + '.')
```

```python
# Example 15: Multi-line docstring detail printing
print(compute_bmi.__doc__)
```

## 💼 Technical Interview Questions & Answers
**Q1: What is the difference between a comment and a docstring in Python?**
*Answer:* Comments are stripped out during bytecode compilation and are ignored at runtime. Docstrings are compiled into Python bytecode and are bound to the `__doc__` attribute of the object (function, class, module), allowing them to be inspected at runtime.

**Q2: How do you read a function's documentation programmatically?**
*Answer:* You can inspect its docstring using `function_name.__doc__` or by calling `help(function_name)` in the shell.

**Q3: Which PEP guidelines cover comments and docstrings?**
*Answer:* PEP 8 covers code styling and comments spacing guidelines. PEP 257 outlines docstring conventions.

**Q4: What is the risk of using multi-line string literals (triple quotes) as comments?**
*Answer:* Stands alone in execution frames, wasting execution memory, and can lead to bugs if placed in nested loop blocks.
