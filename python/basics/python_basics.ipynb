## 🔍 Concept Explanation: Introduction to Python
Python was created by **Guido van Rossum** and first released in **1991**. Its core design philosophy emphasizes **code readability**, using indentation instead of brackets.

### 🌟 Core Features of Python:
1. **Readable and Clean Syntax:** Python reads like plain English.
2. **Dynamically Typed:** You don't declare variable types; Python resolves them at runtime.
3. **Interpreted Language:** Python source code is compiled into intermediate **bytecode** (`.pyc`), which is executed by the **Python Virtual Machine (PVM)**.
4. **Cross-Platform:** The same Python code runs seamlessly on Windows, macOS, and Linux.
5. **Multi-Paradigm:** Supports procedural, Object-Oriented (OOP), and functional programming paradigms.

## 💼 Technical Interview Questions & Answers
**Q1: What is the Python Virtual Machine (PVM)?**
*Answer:* The PVM is the runtime engine of Python. It is an interpreter loop that steps through compiled bytecode instructions, converting them into machine-level code that your CPU executes.

**Q2: Why is Python described as an interpreted language if it has a compilation step?**
*Answer:* Python code is indeed compiled, but it is compiled to intermediate bytecode (`.pyc`), not machine code. This bytecode is then interpreted line-by-line at runtime by the PVM, which is why it retains the flexibility of an interpreted language.

**Q3: Explain dynamic typing.**
*Answer:* Dynamic typing means variable types are bound at runtime rather than compile-time. You do not declare variable types explicitly; the interpreter binds the type of the value to the variable reference dynamically.

**Q4: What is PEP 8?**
*Answer:* PEP 8 is Python's official Style Guide, which defines standard rules for formatting code (such as using 4 spaces for indentation, naming conventions for classes/variables, and line limits) to ensure readability.
