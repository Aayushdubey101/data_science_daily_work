## 🔍 Concept Explanation
Python's `str` class provides built-in methods for text cleaning, case transformations, splitting, joining, and pattern searching.

### 🌟 Why This Matters in Data Science:
Data cleaning routinely begins with removing leading whitespace, standardizing casing, and parsing records into tokens.

### 🧠 Core Concepts & Mental Models:
- **Whitespace removal:** `.strip()`, `.lstrip()`, `.rstrip()`.
- **Casing methods:** `.upper()`, `.lower()`, `.title()`.
- **Split/Join:** `.split(sep)` and `.join(list)`.

---

## 📝 Detailed Explanations
String helper methods return modified string copies without altering the original string.



### 💻 Example 1: Trimming whitespace using strip()


```python
s = '   Data Science   '
print('Original:', repr(s))
print('Stripped:', repr(s.strip()))
```

### 💻 Example 2: Casing methods: upper, lower, title


```python
s = 'data science'
print('Upper:', s.upper())
print('Title:', s.title())
```

### 💻 Example 3: Splitting strings into lists


```python
s = 'apple,banana,cherry'
print('Split list:', s.split(','))
```

### 💻 Example 4: Joining list of strings


```python
lst = ['apple', 'banana', 'cherry']
print('Joined:', ' | '.join(lst))
```

### 💻 Example 5: Replacing substrings


```python
s = 'I love Java'
print('Replaced:', s.replace('Java', 'Python'))
```

### 💻 Example 6: Finding substring indices


```python
s = 'Data Science'
print('Index of "Science":', s.find('Science'))
print('Index of "Java" (missing):', s.find('Java'))
```

## ⚠️ Common Mistakes & Pitfalls
- **Expecting in-place modification:** Assuming `.strip()` modifies the original string; you must reassign the result to persist changes.
- **Calling split() on lists:** Attempting to call `.split()` on lists or other non-string types.

## 💼 Technical Interview Questions & Answers
**Q1: What is the difference between split() and partition()?**
*Answer:* `split()` splits a string into a list of substrings using a separator. `partition()` splits at the first occurrence of a separator and returns a 3-tuple (prefix, separator, suffix).

**Q2: How do you replace all spaces with hyphens in a string?**
*Answer:* Use the `.replace(' ', '-')` method.
