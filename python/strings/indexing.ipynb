## 🔍 Concept Explanation
String indexing allows accessing individual characters using square brackets, supporting both zero-based positive indexing and negative indexing from the end.

### 🌟 Why This Matters in Data Science:
Text parsing, token extraction, and character checks rely on direct index access.

### 🧠 Core Concepts & Mental Models:
- **Positive indexing:** Zero-based, range `0` to `len-1`.
- **Negative indexing:** Ends-based, range `-1` to `-len`.

---

## 📝 Detailed Explanations
Since strings are contiguous arrays of characters in memory, accessing any character by its index is an O(1) constant-time operation.



### 💻 Example 1: Zero-based positive indexing


```python
s = 'DataScience'
print('First character:', s[0])
print('Middle character:', s[4])
```

### 💻 Example 2: Negative indexing from the end of string


```python
s = 'DataScience'
print('Last character s[-1]:', s[-1])
print('Second to last s[-2]:', s[-2])
```

### 💻 Example 3: Handling out-of-bounds IndexError


```python
s = 'Python'
try:
    print(s[10])
except IndexError as e:
    print('IndexError caught:', e)
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Exceeding limits:** Attempting to access indexes outside the range of `len(str)`, raising an IndexError.
- **Negative out-of-bound errors:** Forgetting that negative indexes must be greater than or equal to `-len(str)`.

---

## 💼 Technical Interview Questions & Answers
**Q1: How does negative indexing work?**
*Answer:* It resolves references relative to the end of the sequence, where `-1` represents the last character and `-len(str)` represents the first character.

**Q2: What is the time complexity of accessing a character by its index?**
*Answer:* O(1) constant time, because strings are contiguous arrays in memory.
