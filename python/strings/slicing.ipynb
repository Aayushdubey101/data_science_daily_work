## 🔍 Concept Explanation
String slicing extracts a substring using the formula `string[start:stop:step]`.

### 🌟 Why This Matters in Data Science:
Extracting substrings, reversing text, and stepping through structures are essential parsing skills.

### 🧠 Core Concepts & Mental Models:
- **Inclusive/Exclusive:** Slicing is inclusive of the `start` index and exclusive of the `stop` index.
- **Strides:** The optional `step` parameter controls the increment, allowing reversals with negative strides.

---

## 📝 Detailed Explanations
Slicing creates a new string containing copies of the selected characters. Slicing a substring of length K runs in O(K) time.



### 💻 Example 1: Basic slicing [start:stop]


```python
s = 'PythonProgramming'
print('Slice [0:6]:', s[0:6])
print('Slice [6:13]:', s[6:13])
```

### 💻 Example 2: Slicing with steps [start:stop:step]


```python
s = 'PythonProgramming'
print('Every second character [::2]:', s[::2])
```

### 💻 Example 3: String reversal using negative strides


```python
s = 'Python'
print('Reversed:', s[::-1])
```

### 💻 Example 4: Omitting boundaries (default start and stop)


```python
s = 'Data'
print('Default slice [1:]:', s[1:])
print('Default slice [:3]:', s[:3])
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Off-by-one errors:** Forgetting that the stop index is exclusive.
- **Incompatible negative stride parameters:** Using a negative step with positive bounds (e.g. `s[0:5:-1]` returns an empty string).

---

## 💼 Technical Interview Questions & Answers
**Q1: What is the output of 'hello'[::-1] and how does it work?**
*Answer:* It outputs `'olleh'`. The step value of `-1` reverses the slice progression, stepping backwards from the end.

**Q2: Is the stop index in string slicing inclusive or exclusive?**
*Answer:* Exclusive.
