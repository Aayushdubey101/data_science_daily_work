## 🔍 Concept Explanation
Strings are immutable sequences of Unicode code points. Once a string is created, its characters cannot be modified in place.

### 🌟 Why This Matters in Data Science:
Text processing, NLP pipelines, and data cleaning routines start with robust string manipulation.

### 🧠 Core Concepts & Mental Models:
- **Immutability:** Characters cannot be altered in-place; modifications create new objects.
- **Encoding:** Python strings are natively stored in Unicode format.

---

## 📝 Detailed Explanations
In Python, strings are contiguous arrays of bytes representing Unicode code points. Any operation that changes a character creates a copy of the entire string in memory, running in O(N) time.



### 💻 Example 1: Creating string literals


```python
str1 = 'Hello'
str2 = "World"
print(str1, str2)
```

### 💻 Example 2: Multi-line string representation


```python
paragraph = '''Line A\nLine B'''
print(paragraph)
```

### 💻 Example 3: String immutability representation (reassignment creates new objects)


```python
s = 'Data'
orig_id = id(s)
s += ' Science'
new_id = id(s)
print('Original ID:', orig_id)
print('New ID:', new_id)
print('Are IDs same?', orig_id == new_id)
```

## ⚠️ Common Mistakes & Pitfalls
- **Concatenating strings inside loops:** Using `+` inside a loop creates new string objects in every iteration, running in O(N^2) time. Use `.join()` on lists instead.
- **Expecting string methods to change in-place:** Writing `text.strip()` instead of `text = text.strip()`.

## 💼 Technical Interview Questions & Answers
**Q1: Why are strings immutable in Python?**
*Answer:* Immutability allows strings to be used as dictionary keys (their hash value remains constant) and enables optimization techniques like string interning.

**Q2: What is the time complexity of concatenating N strings in a loop using the '+' operator?**
*Answer:* O(N^2), because a new string is copied in every iteration.
