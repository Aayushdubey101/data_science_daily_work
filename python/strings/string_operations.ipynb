## 🔍 Concept Explanation
Python supports multiple string operations: concatenation (`+`), repetition (`*`), and membership checking (`in`).

### 🌟 Why This Matters in Data Science:
Dynamic query building, dashboard rendering, and log generation require robust formatting.

### 🧠 Core Concepts & Mental Models:
- **Concatenation (+):** Combines strings.
- **Repetition (*):** Repeats strings.
- **Membership (in):** Checks if a substring exists.

---

## 📝 Detailed Explanations
F-strings allow inline float precision styling (e.g. `{val:.2f}`), padding, and alignment, which are resolved at runtime.



### 💻 Example 1: String Concatenation


```python
a = 'Data'
b = 'Science'
print('Concatenation:', a + ' ' + b)
```

### 💻 Example 2: String Repetition


```python
a = 'Python'
print('Repetition:', a * 3)
```

### 💻 Example 3: Membership testing in strings


```python
s = 'Data Science'
print('Data' in s)
print('Java' not in s)
```

### 💻 Example 4: Precision float formatting inside f-strings


```python
val = 3.14159
print(f'Pi to 2 decimal places: {val:.2f}')
```

---

## ⚠️ Common Mistakes & Pitfalls
- **Concatenating mismatched types:** Trying to add an integer directly to a string without explicit casting (e.g., `'Age: ' + 22` raises a `TypeError`).
- **Overcomplicating templates:** Using old-style %-formatting instead of modern, readable f-strings.

## 💼 Technical Interview Questions & Answers
**Q1: How do you verify if a substring exists within a string?**
*Answer:* Use the `in` operator (e.g. `'sub' in 'string'`).

**Q2: What is the output of 'Python' * 3?**
*Answer:* `'PythonPythonPython'`.
