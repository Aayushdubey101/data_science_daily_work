```python
# Convert generator expression to tuple
numbers = tuple(x**2 for x in range(5))
print(numbers)  # (0, 1, 4, 9, 16)
# Parentheses create generator, not tuple
```

```python
# This creates a generator object, not a tuple
gen = (x**2 for x in range(5))
print(type(gen))  # <class 'generator'>

# Convert to tuple
result = tuple(gen)
print(result)  # (0, 1, 4, 9, 16)
print(type(result))  # (0, 1, 4, 9, 16)
```

```python
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4, 5))  # 15
print(sum_all(10, 20))          # 30
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30, city="New York")
# name: Alice
# age: 30
# city: New York
```

```python
def calculate(a, b, c, operation="add"):
    if operation == "add":
        return a + b + c
    elif operation == "multiply":
        return a * b * c

# Unpacking list/tuple with *
numbers = [5, 10, 15]
result1 = calculate(*numbers)  # Same as calculate(5, 10, 15)

# Unpacking dict with **
params = {"a": 2, "b": 3, "c": 4, "operation": "multiply"}
result2 = calculate(**params)  # Same as calculate(a=2, b=3, c=4, operation="multiply")

print(result1)  # 30
print(result2)  # 24
```

```python
# use of map function
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
squared
```
