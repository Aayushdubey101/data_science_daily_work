```python
# Rolling window
```

```python
df['Close'].rolling(3)
```

```python
# Plot rolling mean (orange) first, then original Close (blue) on top so Close remains visible
plt.figure(figsize=(12,6))
# rolling mean plotted with some transparency and lower zorder
df['Close'].rolling(150).mean().plot(color='orange', linewidth=2, alpha=1, label='3-period Rolling Mean', zorder=2)
# original Close plotted on top with higher zorder
df['Close'].plot(color='blue', linewidth=2.2, label='Close', zorder=1)
plt.title('GOOGL Close Price and 3-period Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Plot rolling mean (orange) first, then original Close (blue) on top so Close remains visible
plt.figure(figsize=(12,6))
# rolling mean plotted with some transparency and lower zorder
df['Close'].ewm(15).mean().plot(color='orange', linewidth=2, alpha=1, label='5 EWM', zorder=2)

df['Close'].rolling(150).mean().plot(color='green', linewidth=2, alpha=1, label='3-period Rolling Mean', zorder=2)
# original Close plotted on top with higher zorder
df['Close'].plot(color='blue', linewidth=2.2, label='Close', zorder=1)
plt.title('GOOGL Close Price ')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(alpha=0.4)
plt.tight_layout()
plt.show()
```
