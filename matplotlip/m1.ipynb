```python

labels = x["batsman"]
y2015 = x["2015"]
y2016 = x["2016"]
y2017 = x["2017"]

pos = np.arange(len(labels))
width = 0.25

plt.figure(figsize=(12, 5))
plt.bar(pos - width, y2015, width=width, label="batsman runs in 2015")
plt.bar(pos,         y2016, width=width, label="batsman runs in 2016", color="red")
plt.bar(pos + width, y2017, width=width, label="batsman runs in 2017")

plt.xticks(pos, labels, rotation=45)
plt.xlabel("Batsman")
plt.ylabel("Runs")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# stacked bar plot
labels = x["batsman"]
y2015 = x["2015"]
y2016 = x["2016"]
y2017 = x["2017"]
pos = np.arange(len(labels))
plt.figure(figsize=(12, 5))
plt.bar(pos, y2015, label="batsman runs in 2015")
plt.bar(pos, y2016, bottom=y2015, label="batsman runs in 2016", color="red")
plt.bar(pos, y2017, bottom=y2015+y2016, label="batsman runs in 2017")
```

```python
plt.bar(x["batsman"],x["2015"],label="2015")
plt.bar(x["batsman"],x["2016"],label="2016",bottom=x["2015"])
plt.bar(x["batsman"],x["2016"],label="2016",bottom=(x["2015"]+x["2016"]))
plt.legend()
plt.show()
```
