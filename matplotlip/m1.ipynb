```python
batsman.rename(columns={"index":"session"}, inplace=True)
```

```python
plt.plot(batsman["session"], batsman["V Kohli"])
```

```python
# multiple line plot
plt.plot(batsman["session"], batsman["V Kohli"], label="V Kohli")
plt.plot(batsman["session"], batsman["RG Sharma"], label="R Sharma")
# add labels and title
plt.xlabel("Session")
plt.ylabel("Runs")
plt.title("Batsman Performance Across Sessions")
plt.legend()
```

```python
# use linewidth, linestyle and marker
plt.plot(batsman["session"], batsman["V Kohli"], label="V Kohli", linewidth=2, linestyle='--', marker='o',color='Red')
plt.plot(batsman["session"], batsman["RG Sharma"], label="R Sharma", linewidth=3, linestyle='-.', marker='d',markersize=8,color='Blue')
# add labels and title
plt.xlabel("Session")
plt.ylabel("Runs")
plt.title("Batsman Performance Across Sessions")
plt.legend()
```

```python
# handling outliers
batsman.loc[5, "V Kohli"] = 15000
plt.plot(batsman["session"], batsman["V Kohli"], label="V Kohli", linewidth=2, linestyle='--', marker='o',color='Red')
plt.plot(batsman["session"], batsman["RG Sharma"], label="R Sharma", linewidth=3, linestyle='-.', marker='d',markersize=8,color='Blue')
# add labels and title
plt.xlabel("Session")
plt.ylabel("Runs")
plt.title("Batsman Performance Across Sessions")
plt.legend()
```
