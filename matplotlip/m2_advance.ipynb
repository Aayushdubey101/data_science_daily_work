```python
plt.figure(figsize=(18,10))
plt.scatter(batts['avg'], batts['strike_rate'], s=batts['runs'])
for idx, row in batts.iterrows():
    plt.text(row['avg'], row['strike_rate'], row['batter'],)
```

```python
# example of 3d line plot
x=np.linspace(0, 10, 100)
y=np.random.randint(0, 10, 100)
z=np.random.randint(0, 10, 100)

fig =plt.figure()
ax=fig.add_subplot(111, projection='3d',)
ax.plot3D(x,y,z)
ax.set_xlabel('avg')
ax.set_ylabel('runs')
ax.set_zlabel('strike_rate')
ax.set_title('3D Line Plot')
```

```python
# 3d surface plot
x=np.linspace(-5, 5, 100)
y=np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z=np.sin(np.sqrt(x**2 + y**2))
fig =plt.figure(figsize=(15,15))
ax=fig.add_subplot(111, projection='3d',)
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
#color bar
fig.colorbar(ax.plot_surface(x, y, z, cmap='viridis'), shrink=0.5, aspect=5)

ax.set_title('3D Surface Plot')
```

```python
# in 3d
x=np.linspace(-5, 5, 100)
y=np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z=x**2 + y**2
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111, projection='3d',)

ax.plot_surface(x, y, z, cmap='viridis')
fig.colorbar(ax.plot_surface(x, y, z, cmap='viridis'), shrink=0.5, aspect=5)
```

```python
# contur plot
x=np.linspace(-5, 5, 100)
y=np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z=x**2 + y**2
plt.figure(figsize=(10,10))
plt.contour(x, y, z, levels=20, cmap='viridis')
plt.colorbar()
```
